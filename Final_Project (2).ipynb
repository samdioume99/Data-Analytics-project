```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt 

Data = "https://raw.githubusercontent.com/samdioume99/nba_stats/main/totals_stats.csv"
df = pd.read_csv(Data)
print(df)
lebrondf = df[df['Player'] == 'Lebron James']
print("Lebron Avg total points per Season-", lebrondf['PTS'].mean())
print("Lebron Avg total assists per Season-", lebrondf['AST'].mean())
print("Lebron Avg total blocks per Season-", lebrondf['BLK'].mean())
print("Lebron Avg total steals per Season-", lebrondf['STL'].mean())
print("Lebron Avg total FG% per Season-", lebrondf['FG%'].mean())
print()


Jordandf = df[df['Player'] == 'Michael Jordan']
print("Jordan Avg total points per Season-", Jordandf['PTS'].mean())
print("Jordan Avg total assists per Season-", Jordandf['AST'].mean())
print("Jordan Avg total blocks per Season-", Jordandf['BLK'].mean())
print("Jordan Avg total steals per Season-", Jordandf['STL'].mean())
print("Jordan Avg total FG% per Season-", Jordandf['FG%'].mean())
print()

Kobedf = df[df['Player'] == 'Kobe Bryant']
print("Kobe Avg total points per Season-", Kobedf['PTS'].mean())
print("Kobe Avg total assists per Season-", Kobedf['AST'].mean())
print("Kobe Avg total blocks per Season-", Kobedf['BLK'].mean())
print("Kobe Avg total steals per Season-", Kobedf['STL'].mean())
print("Kobe Avg total FG% per Season-", Kobedf['FG%'].mean())
print()

lebrondf = df[df['Player'] == 'Lebron James']
print("Lebron median total points per Season-", lebrondf['PTS'].median())
print("Lebron median total assists per Season-", lebrondf['AST'].median())
print("Lebron median total blocks per Season-", lebrondf['BLK'].median())
print("Lebron median total steals per Season-", lebrondf['STL'].median())
print("Lebron median total FG% per Season-", lebrondf['FG%'].median())
print()

Jordandf = df[df['Player'] == 'Michael Jordan']
print("Jordan median total points per Season-", Jordandf['PTS'].median())
print("Jordan median total assists per Season-", Jordandf['AST'].median())
print("Jordan median total blocks per Season-", Jordandf['BLK'].median())
print("Jordan median total steals per Season-", Jordandf['STL'].median())
print("Jordan median total FG% per Season-", Jordandf['FG%'].median())
print()

Kobedf = df[df['Player'] == 'Kobe Bryant']
print("Kobe Avg median points per Season-", Kobedf['PTS'].median())
print("Kobe Avg median assists per Season-", Kobedf['AST'].mean())
print("Kobe Avg median blocks per Season-", Kobedf['BLK'].mean())
print("Kobe Avg median steals per Season-", Kobedf['STL'].mean())
print("Kobe Avg median FG% per Season-", Kobedf['FG%'].mean())
print()

print("Each number is unique so mode is N/A")

x = range(19,34)
y = df.head(15)["PTS"]

plt.plot(x, y)

plt.xlabel('AGE')
#plt.ylabel('PTS')
plt.title('Lebron James Regular Season Points Comparison')

plt.show()

df = df.iloc[31:46, :]

# Plot the points
plt.plot(df["PTS"])

# Add axis labels and a title
plt.xlabel("AGE")
#plt.ylabel("PTS")
plt.title("Michael Jordan Regular Season Points Comparison")

# Show the plot
plt.show()


df = df.iloc[59:78, :]

x = range(18,37)
y = df.iloc[59:78, :]

# Plot the points
plt.plot(df["PTS"])
#plt.plot(x, y) 

# Add axis labels and a title
plt.xlabel("AGE")
#plt.ylabel("PTS")
plt.title("Kobe Bryant Regular Season Points Comparison")

# Show the plot
plt.show()

section1 = df.loc[2:29, "PTS"]
section2 = df.loc[31:58, "PTS"]
section3 = df.loc[59:93, "PTS"]

# Create a pie chart
plt.pie([section1.sum(), section2.sum(), section3.sum()], labels=["Lebron", "Jordan", "Section 3"])
plt.show()
```
